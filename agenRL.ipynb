```python
import gym 
from gym import Env
from gym.spaces import Discrete, Box, Dict, Tuple, MultiBinary, MultiDiscrete 
import numpy as np
import random
import os
#stable-baseline library for RL
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import VecFrameStack
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
from stable_baselines3.common.env_checker import check_env
```

```python
from train import GameOf2048#import logic game with rewards and penalties
```

```python
env = GameOf2048()
```

```python
#test env for 2048
episodes = 5
for episode in range(1, episodes+1):
    state = env.reset()
    done = False
    score = 0
    cont = 0
    
    while not done:
        #env.render(0)
        action = env.action_space.sample()
        n_state, reward, done, info = env.step(action)
        score+=reward
        cont += 1
    print('Episode:{} Score:{}, steps {}'.format(episode, score, cont))
env.close()
```

```python
log_path = os.path.join('Training', 'Logs')
```

```python
#set 4 layers of 128 neurons
net_arch=[dict(pi=[128, 128, 128, 128], vf=[128, 128, 128, 128])]
```

```python
#set ppo model 
model = PPO('MlpPolicy', env, verbose = 1, policy_kwargs={'net_arch': net_arch})
```

```python
#trains model 1000000 times
#reward: poins earned in game
#penalties: trash movemnts
#           non-maximal movments
model.learn(total_timesteps=1000000)
```

```python
model.save('PPO')#save model
```

```python
model = PPO.load('PPO')#load model
```

```python
from game import GameOf2048 as Eval2048
env = Eval2048()
#evalueate model
evaluate_policy(model, env, n_eval_episodes=100, render=False)
#prints avg score and std score
```
